### 4673
https://www.acmicpc.net/problem/4673
image.png

```python
# n을 d(n)의 생성자
# 33은 39의 생성자  33 -> 39
# 생성자가 없는 숫자를 셀프 넘버라고 함
a = set()
total_num = set(range(1,10001))
for i in range(1,10001):
    for j in str(i):
        i += int(j)
    a.add(i)
self_num = sorted(total_num-a)
for i in self_num:
    print(i) 
```

### 1065
https://www.acmicpc.net/problem/1065
image.png

```python
N = int(input())
cnt = 0
for i in range(1,N+1):
    if i <= 99:
        cnt += 1
    else:
        a = list(map(int,str(i)))
        if a[0] - a[1] == a[1] - a[2]:
            cnt += 1
print(cnt)
```
