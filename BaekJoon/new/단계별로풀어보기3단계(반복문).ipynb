### 1110
https://www.acmicpc.net/problem/1110
image.png

```python
N = int(input())

num = N
cnt = 0
while(1):
    a = num // 10
    b = num % 10
    c = (a+b)% 10
    num = b*10 + c
    cnt += 1
    if num == N:
        break
print(cnt)
```
