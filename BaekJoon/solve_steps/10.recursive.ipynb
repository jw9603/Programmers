```python
#1.
import sys
N = int(sys.stdin.readline())
def factorial(n):
    if n==0:
        return 1
    elif n==1:
        return 1
    return factorial(n-1)*n
print(factorial(N))
```

```python
#2
def fibo(n):
    if n==0:
        return 0
    elif n==1:
        return 1
    else:
        return fibo(n-1)+fibo(n-2)
n = int(input())
print(fibo(n))
```

```python
#2. for문 사용(dynamic programming)
n = int(input())
fibo=[0,1]
for i in range(2,n+1):
    num = fibo[i-1]+fibo[i-2]
    fibo.append(num)
print(fibo[n])
```

```python
#3
N = int(input())

stars = [[' ' for _ in range(N)] for _ in range(N)]

def fill_star(size,x,y):
    if size == 1:
        stars[y][x]='*'
    else:
        next_size = size//3
        for dy in range(3):
            for dx in range(3):
                if dy!=1 or dx!=1:
                    fill_star(next_size,x+dx*next_size,y+dy*next_size)
                    
fill_star(N,0,0)
for k in stars:
    print(''.join(k))
```

```python
#3
def star(n:int,x:list):
    out=[]
    if n==3:
        return x
    else:
        for i in x:
            out.append(i*3)
        for i in x:
            out.append(i+' '*len(x)+i)
        for i in x:
            out.append(i*3)
        return star(n//3,out)
        #크기가 N이라면 그 전단계인 N/3인 정사각형을 이용하여 배열하게 됨
n = int(input())
first=['***','* *','***']
final = star(n,first)
for i in final:
    print(i)
```
